```python
# Import all the necessary libraries

import numpy as np
import pandas as pd
from datetime import datetime as dt
import itertools

%matplotlib inline
```

```python
# Read data from the CSV into a dataframe

raw_data_1 = pd.read_csv('resource\\2007-08.csv')
raw_data_2 = pd.read_csv('resource\\2008-09.csv')
raw_data_3 = pd.read_csv('resource\\2009-10.csv')
raw_data_4 = pd.read_csv('resource\\2010-11.csv')
raw_data_5 = pd.read_csv('resource\\2011-12.csv')
raw_data_6 = pd.read_csv('resource\\2012-13.csv')
raw_data_7 = pd.read_csv('resource\\2013-14.csv')
raw_data_8 = pd.read_csv('resource\\2014-15.csv')
raw_data_9 = pd.read_csv('resource\\2015-16.csv')
raw_data_10 = pd.read_csv('resource\\2016-17.csv')
raw_data_11 = pd.read_csv('resource\\2017-18.csv')
raw_data_12 = pd.read_csv('resource\\2018-19.csv')
raw_data_13 = pd.read_csv('resource\\2019-20.csv')
raw_data_14 = pd.read_csv('resource\\2020-21.csv')
raw_data_15 = pd.read_csv('resource\\2021-22.csv')
```

```python
# Parse data as time 

def parse_date(date):
    if date == '':
        return None
    else:
        return dt.strptime(date, '%d/%m/%y').date()
    

def parse_date_other(date):
    if date == '':
        return None
    else:
        return dt.strptime(date, '%d/%m/%Y').date()

raw_data_1.Date = raw_data_1.Date.apply(parse_date)    
raw_data_2.Date = raw_data_2.Date.apply(parse_date)    
raw_data_3.Date = raw_data_3.Date.apply(parse_date)         # The date format for this dataset is different  
raw_data_4.Date = raw_data_4.Date.apply(parse_date)    
raw_data_5.Date = raw_data_5.Date.apply(parse_date)    
raw_data_6.Date = raw_data_6.Date.apply(parse_date)    
raw_data_7.Date = raw_data_7.Date.apply(parse_date)    
raw_data_8.Date = raw_data_8.Date.apply(parse_date)    
raw_data_9.Date = raw_data_9.Date.apply(parse_date_other)    
raw_data_10.Date = raw_data_10.Date.apply(parse_date)
raw_data_11.Date = raw_data_11.Date.apply(parse_date_other)
raw_data_12.Date = raw_data_12.Date.apply(parse_date_other)
raw_data_13.Date = raw_data_13.Date.apply(parse_date_other)
raw_data_14.Date = raw_data_14.Date.apply(parse_date_other)
raw_data_15.Date = raw_data_15.Date.apply(parse_date_other)
```

```python
#Gets all the statistics related to gameplay
                      
playing_statistics_1 = raw_data_1                     
playing_statistics_2 = raw_data_2
playing_statistics_3 = raw_data_3
playing_statistics_4 = raw_data_4
playing_statistics_5 = raw_data_5
playing_statistics_6 = raw_data_6
playing_statistics_7 = raw_data_7
playing_statistics_8 = raw_data_8
playing_statistics_9 = raw_data_9
playing_statistics_10 = raw_data_10
playing_statistics_11 = raw_data_11
playing_statistics_12 = raw_data_12
playing_statistics_13 = raw_data_13
playing_statistics_14 = raw_data_14
playing_statistics_15 = raw_data_15
```

FINAL DATAFRAME

```python
playing_stat = pd.concat([playing_statistics_1,
                          playing_statistics_2,
                          playing_statistics_3,
                          playing_statistics_4,
                          playing_statistics_5,
                          playing_statistics_6,
                          playing_statistics_7,
                          playing_statistics_8,
                          playing_statistics_9,
                          playing_statistics_10,
                          playing_statistics_11,
                          playing_statistics_12,
                          playing_statistics_13,
                          playing_statistics_14,
                          playing_statistics_15],                          
                         ignore_index = True)


```
